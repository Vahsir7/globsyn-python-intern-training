```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### PLotting

```python
x = np.linspace(1,50)
y1 = np.array((2*x)+1)
y2 = np.array(x**2+10)
y3 = np.array(np.log2(x))
y4 = np.array(2**x)
```

```python
plt.title("log func")
plt.xlabel("x")
plt.ylabel("y = log2(x)")
plt.plot(x,y3)
```

### Saving Graphs

```python
plt.figure(figsize=(12,5))
plt.xticks(np.arange(1,50,3))
plt.plot(x,y2)
plt.savefig(r'E:\Globsyn\imagess\img.png')
```

### Customising PLots

```python
plt.figure(figsize=(15,5))
plt.plot(x,y2,marker = 'o', mfc = 'y', mec = 'c',ls=":")
```

* subploting

```python
plt.subplot(221)
plt.plot(x,y1)
plt.subplot(222)
plt.plot(x,y2)
plt.subplot(223)
plt.plot(x,y3)
plt.subplot(224)
plt.plot(x,y4)
plt.tight_layout()
```

* All in one plot

```python
plt.figure()
plt.plot(x,y1,ls=":",label="g1")
plt.plot(x,y2,label = "g2",c="k")
plt.plot(x,y3, label = "g3")
plt.legend()
plt.show()
```

```python
u = np.random.normal(1.5,3,1000)
print(type(u))
plt.figure()
plt.hist(u,color='k',edgecolor='y',bins=10)
plt.show()
```

```python
plt.boxplot(u,showmeans = True)
plt.show()
```

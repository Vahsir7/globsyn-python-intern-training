##### row selection
- loc - it is location function will take a **list** of index names and column names 
- iloc - it is index based selection from dataframe

in iloc we can pass list or slicing list

```python
print(df.loc[[50,60],['pl','pw','sp']])
print('\n\n')
print(df.iloc[50:61:2,[0,-1]])
print('\n\n')
print(df.iloc[[1] + list(range(50, 61, 2)) + [df.index[-1]],[0,-1]])
```

### Conditional selection
select the rows with sepal length more than the mean sepal length

```python
sl_avg=df['sl'].mean()
sl= df['sl']
selected_rows=df[sl>sl_avg]
print(selected_rows[['sl','sp']])
```

we are dropping the categorical value and finding the correlation among them

```python
df1=df.drop('sp',axis=1)
print(df1)

df2=df1.corr()
print(df2)

```

#### Categorical data manipulation


>'categorical column'.function()

-   e,g sepal.value_counts gives count for each value

- nunique - gives the number of values
- unique - gives the names of unique values

```python
print(df.sp.value_counts())

print("\nUnique values:",df.sp.nunique())
print(df.sp.unique())

```

```python
print("\n\n SORTING BY AXIS 0")
print(df.sort_values(by='sl'))

print("\n\n SORTING BY AXIS 1")
print(df.sort_index())
```

replace function ( [intial categories] ,[replaced categories] )

```python
df1=df.replace(['setosa','versicolor','virginica'],[0,1,2])
print(df1)
print("\n\n")

df1=df.replace({'setosa':0,'versicolor':1,'virginica':2})#will be replaced#
print(df1)
```

group by -

```python
df1=df.groupby('sp').mean()
print(df1)
```

```python
df1=df.groupby('sp').agg(['mean','std'])
print(df1)

print("\n\n")

df1=df.groupby('sp').agg({'sl':'mean','sw':'std'})
print(df1)
```
